```python
data['login_date'] =  pd.to_datetime(data['login_date'], format='%m/%d/%y') #login_date format changing into another date format for better data analysis

data['day_of_week'] = data['login_date'].dt.day_name() #to extract and view the days of a week from date format
```

```python
column1=data[(data['projects_added']==True)]

column2=data[(data['likes_given']==True)]

column3=data[(data['comment_given']==True)]

column4=data[(data['day_of_week']==True)]

column5 = data[(data['bug_occured']==True)]
```

```python
#to create a data table, customer_id and day_of week columns are already added. Now, there will be three new columns in this new data table.
#count_no_of_projects column is the column which counts the number of projects added by customer
#count_no_of_likes column is the column which counts the number of likes made by customer
#count_no_of_comments is the column which counts the number of comments made by customer
column11=column1.groupby(['customer_id','day_of_week'])['projects_added'].count().reset_index(name="count_no_of_projects") 

column12=column2.groupby(['customer_id','day_of_week'])['likes_given'].count().reset_index(name="count_no_of_likes")

column13=column3.groupby(['customer_id','day_of_week'])['comment_given'].count().reset_index(name="count_no_of_comments")
```

```python
data1=pd.concat([column11,column12,column13,column4], axis=1)

data1=data1.iloc[:,[0,1,2,5,8]]
```

```python
data1 #to view the new data table
```

```python
#to analyze the user details like how much active user is on showwcase website
data1['active_user']=data1['count_no_of_projects']+data1['count_no_of_likes']+data1['count_no_of_comments']
```
